prep structures for tfrecord writing and write tfrecords

```python
def bytes_feature(value):
        
    return tf.train.Feature(bytes_list=tf.train.BytesList(value=[value]))
 
def int64_feature(value):

    return tf.train.Feature(int64_list=tf.train.Int64List(value=[value]))



def convert_to(data_set, name):
  """Converts a dataset to tfrecords."""

    
  num_examples = len(data_set)
  if not os.path.exists(name):
     os.makedirs(name)
#   if images.shape[0] != num_examples:
#     raise ValueError('Images size %d does not match label size %d.' %
#                      (images.shape[0], num_examples))
  rows = data_set[0][1].shape[0]
  cols = data_set[0][1].shape[1]
  depth = data_set[0][1].shape[2]

  filename = (name + '.tfrecords')
  print('Writing', filename)
  writer = tf.python_io.TFRecordWriter(filename)
  for index in range(num_examples):
    image_raw = data_set[index][1].tostring()
    label_raw = data_set[index][0].tostring()
    example = tf.train.Example(features=tf.train.Features(feature={
        'height': int64_feature(rows),
        'width': int64_feature(cols),
        'depth': int64_feature(depth),
        'label': bytes_feature(label_raw),
        'image_raw': bytes_feature(image_raw)}))
    writer.write(example.SerializeToString())
  writer.close()
```
